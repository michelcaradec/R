# Circlize Library

```r
product.count <- 5
products <- paste0("Product ", letters[1:product.count])
```

List of available products:

```r
products
```

## Using a dataframe

```r
count <- 100
df <- data.frame(owns = sample(products, count, replace = T), buys = sample(products, count, replace = T))
df <- df[df$owns != df$buys, ] # Remove A => A rows
head(df)
```

Customers who own a product usually buy other products.

```r
library(circlize)
chordDiagram(df)
```

Product links are displayed one-by-one (width = 1).

## Using a matrix

```r
library(reshape2)
(df.m <- suppressMessages(acast(data = df, formula = buys ~ owns)))
```

```r
chordDiagram(df.m)
```

Product links are grouped then displayed (width = number of (product 1, product 2) tuple occurences).
